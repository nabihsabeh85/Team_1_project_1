```python
# import relevent liberaries to create graphs
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import seaborn as sns
```

```python
#Use Kaggle to source data
#https://www.kaggle.com/datasets/jacksondivakarr/sample34
#create the Data Frame (car_data_df)

car_data = 'processes2.csv'
car_data_df = pd.read_csv('processes2.csv')
car_data_df
```

```python
#Clean data frame (remove irrelevent columns)
car_data_df.drop("Unnamed: 0", axis=1 , inplace = True )
car_data_df
```

```python
# drop additional columns that are not relevent to the problem we are trying to solve ("Mileage Unit", "Engine (CC))

car_data_df = car_data_df.drop(columns=["Mileage Unit", "Engine (CC)"])
car_data_df
```

```python
#rename columns for clear representation

car_data_df = car_data_df.rename(
    columns = {
        "name": "Make", "year": "Year", "selling_price": "Price", "km_driven": "Odometer" , "fuel": "Fuel","seller_type": "Seller Type",
    "transmission": "Transmission","owner": "Owner","seats": "Seats","max_power (in bph)": "BPH"
   }
)

car_data_df
```

```python
#use Groupby and mean to get the average price of the cars per Make

avg_price_per_make = car_data_df.groupby('Make')['Price'].mean()
```

```python
avg_price_per_make = avg_price_per_make.reset_index()
```

```python
avg_price_per_make.plot(kind='bar', x='Make', y='Price')
#use plot to create a bar graph
```

```python
#Create Scatter plot graph to show the relationship between the selling price and year of Manufacture

plt.figure(figsize=(10, 6))
plt.ticklabel_format(useOffset=False)
sns.scatterplot(x='Year', y='Price', data=car_data_df, alpha=0.6)
plt.title('Price vs. Year of Manufacture')
plt.xlabel('Year of Manufacture')
plt.ylabel('Selling Price (in INR)')
plt.grid(True)
plt.show()
```

```python
#

plt.figure(figsize=(10, 6))
sns.histplot(car_data_df['Odometer'], bins=30)
plt.title('Histogram of Kilometers Driven Before Selling')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel', y='Price', data=car_data_df)
plt.title('Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price (in INR)')
plt.grid(True)
plt.show()
```

```python
transmission_counts = car_data_df['Transmission'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cars by Transmission Type')

# Show the plot
plt.show()
```

```python


# Assuming 'car_data_df' is your DataFrame and it contains a column named 'Owner'
# First, make a copy to avoid modifying the original DataFrame
car_data_copy = car_data_df.copy()

# Consolidate all owners beyond the second owner into a single category
car_data_copy['Owner'] = car_data_copy['Owner'].apply(lambda x: x if x in ['First Owner', 'Second Owner'] else 'Third or More Owners')

# Calculate the counts of each owner category
owner_counts_consolidated = car_data_copy['Owner'].value_counts()

# Create the pie chart with the consolidated owner counts
plt.figure(figsize=(8, 8))
plt.pie(owner_counts_consolidated, labels=owner_counts_consolidated.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Cars by Owner Type')

# Show the plot
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(car_data_df['Mileage'], kde=True, bins=30)
plt.title('Mileage Efficiency Distribution')
plt.xlabel('Mileage (kmpl)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BPH', y='Price', data=car_data_df, alpha=0.6)
plt.title('Max Power vs. Selling Price')
plt.xlabel('Max Power (BHP)')
plt.ylabel('Selling Price (in INR)')
plt.grid(True)
plt.show()
```

```python
current_year = 2024
car_data_df['Car Age'] = current_year - car_data_df['Year']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='Odometer', data=car_data_df, alpha=0.6)
plt.title('Car Age vs. Kilometers Driven')
plt.xlabel('Car Age (Years)')
plt.ylabel('Kilometers Driven')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel', y='Odometer', data=car_data_df)
plt.title('Fuel Type vs. Mileage Efficiency')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage (kmpl)')
plt.grid(True)
plt.show()
```

```python

# Assuming original_data is your DataFrame containing the car data

# Bin the odometer readings into intervals
bin_edges = np.linspace(car_data_df['Odometer'].min(), car_data_df['Odometer'].max(), num=20)
bin_labels = (bin_edges[:-1] + bin_edges[1:]) / 2  # Calculate bin centers

# Categorize each car into a bin based on odometer reading
car_data_df['Odometer_bin'] = pd.cut(car_data_df['Odometer'], bins=bin_edges, labels=bin_labels)

# Calculate the average price for each bin
avg_price_per_bin = car_data_df.groupby('Odometer_bin')['Price'].mean()

# Plot the averages as a line graph
plt.figure(figsize=(12, 8))
plt.plot(avg_price_per_bin.index.astype(float), avg_price_per_bin.values, marker='o', linestyle='-', color='red')
plt.title('Average Car Price by Odometer Reading')
plt.xlabel('Odometer Reading (km)')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```
